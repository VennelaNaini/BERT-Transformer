```python
from transformers import BertTokenizer, TFBertForSequenceClassification
import tensorflow as tf
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased')
text = "I did not love this product!"
inputs = tokenizer(text, return_tensors='tf', truncation=True, padding=True, max_length=512)
outputs = model(inputs)
logits = outputs.logits
probabilities = tf.nn.softmax(logits, axis=-1)
predicted_class = tf.argmax(probabilities, axis=-1).numpy()[0]
print("Predicted Sentiment:", "Positive" if predicted_class == 1 else "Negative")
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
import tensorflow as tf
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
model = TFAutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
text = """It started out as the perfect day. I woke up early to the sound of birds chirping and the smell of fresh coffee.
The sunlight streamed through the curtains, and everything felt calm and full of promise. I got dressed,
left the house with a smile, and walked to the store to buy ingredients for my favorite dish. On the way, I helped
a stranger pick up their fallen groceries, and we shared a laugh. The weather was ideal — not too warm,
not too cold — and the flowers in the park were in full bloom. But then, out of nowhere, it all changed.
The sky grew dark. I received a call from the hospital that my dog had been hit by a car.
The news hit me like a freight train. I rushed to the hospital only to be told that he didn’t make it.
I sat in the corner, silent, broken, wondering how a beautiful day could turn into the worst one of my life in a matter of hours.
Even now, as I write this, I feel that pain fresh in my chest, and it’s something I’ll carry forever."""
def split_into_chunks(text, max_tokens=512, overlap=50):
    tokens = tokenizer.tokenize(text)
    chunks = []
    for i in range(0, len(tokens), max_tokens - overlap):
        chunk = tokens[i:i + max_tokens]
        chunk_text = tokenizer.convert_tokens_to_string(chunk)
        chunks.append(chunk_text)
    return chunks
chunks = split_into_chunks(text)
sentiments = []

for chunk in chunks:
    inputs = tokenizer(chunk, return_tensors='tf', truncation=True, padding=True, max_length=512)
    outputs = model(inputs)
    logits = outputs.logits
    probs = tf.nn.softmax(logits, axis=-1)
    pred = tf.argmax(probs, axis=-1).numpy()[0]
    sentiments.append(pred)
final_sentiment = max(set(sentiments), key=sentiments.count)
print("Final Predicted Sentiment:", "Positive" if final_sentiment == 1 else "Negative")
```
